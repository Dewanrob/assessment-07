```python
import requests
from bs4 import BeautifulSoup
import datetime
import re
import json
```

```python
url = 'https://www.timeanddate.com/holidays/us/2020'
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

```python
# This gets all the dates
test_holidays = soup.find_all('a', href = re.compile('holidays/us'))
```

```python
holiday_url = url + '/' + test_holidays[0]["href"]
```

```python
holiday_url
```

```python
test_three = soup.find_all(['td', 'th'])
```

```python
test_three
```

```python
# This gets Date, Holiday, Type, and States
for i in range(0, len(test_three)):
    try:
        print(test_three[i-2].get_text(), 
        test_three[i].find('a', href = re.compile('holidays/us')).get_text(),
        test_three[i+1].get_text(),
        test_three[i+2].get_text()
        )
    except:
        pass
```

```python
# Here's a giant dictionary of the holidays
results = []
for i in range(0, len(test_three)):
    try:
        results.append(
            { 
            'Date'      : test_three[i-2].get_text(), 
            'Holiday'   : test_three[i].find('a', href = re.compile('holidays/us')).get_text(),
            'Type'      : test_three[i+1].get_text(),
            'Details'   : test_three[i+2].get_text().replace('\xa0', '')
            }   
        )
    except:
        pass
print(results)
```
